```python
from sportsipy.nba.teams import Teams
from sportsipy.nba.roster import Roster
from sportsipy.nba.roster import Player
```

```python
arr = []
for i in range(1950, 2022):
    teams = Teams(i)
    for team in teams:
        print(i, team.name)
        roster = team.roster
        for player in roster.players:
            arr.append([i, team.name, team.abbreviation, player.name, player.player_id])
            
```

```python
print(len(arr))
```

```python
import csv 
import numpy as np
file = open('lmao.csv', 'w+', newline ='') 
with file:     
    write = csv.writer(file) 
    write.writerows(arr) 
```
